#### **According to the authors, what are the differences between Violent extremism or non-violent extremism?**
Violent extremism must constitute a violent action, or a threat of so. Non-violent extremism does not constitute a threat of violence nor violence itself.
Violent extremists tend to have a low self-esteem compared to NVEs.
Violent extemists also were found by the authors to have been exposed to extreme violence including through the internet or bullying. NVEs do not share the same experience.

#### **Based on the paper, how is radicalization and extremism defined?**
The authors argue that radicalization, which involve having different opinions from the mainstream is one of the pathways to extremism-linked violence. However, they note that the majority of people with radical ideas are not involved in violence, and that radicalization is not the only way to extremism.
#### **How is violent and non-violent extremism defined?**
The authors describe violent extremism as acts or propositions that, if inacted, would result in murder, attempted murder, manslaughter or serious injury to another person or people. This definition expands to involve people who conduct non-violent behaviors that support violence.

On the other hand, the authors left their definition of non-violent extremism to user inference, in which we identified as expression of ideas that diverge from mainstream but do not constitute a threat of action or the action itself. The individuals they collected data from were all convicted of criminal charges linked to extremism of some kind.

````python
class gpt_model:
    @staticmethod
    def prompt(text, model_name = "azure/gpt-4.1-mini"):
      prompt = f"""
        You are a research assistant applying the classification framework from the 2017 Dstl study (Knight et al.) on extremist types.
        Analyze the provided message and classify it as Violent Extremist (VE) or Non-Violent Extremist (NVE).
        Base your classification strictly on the following definitions from Knight, Woodward, and Lancaster (2017):
        1. Classification Criteria:
            Violent Extremist (VE): The message contains a "serious threat of action," advocates for "structural damage," expresses support for physical attacks, or indicates a "personal responsibility to act"
            Non-Violent Extremist (NVE): The message expresses "vocal opposition" to mainstream values (e.g., democracy, rule of law) or "ingroup-outgroup" grievances (e.g., "The West hates us") but does not mention physical force, damage, or specific threats.
            Look for "Cognitive Openings": Does the post suggest a "catalyst" or "crisis" that makes the user feel they must act?

        2. Output:
            For every message input, return its classification label: VE or NVE.
            For VE posts, subgroup them into three subtypes: ideological, political, religious.
            For each post, identify the target group. Focus on demographics e.g
                Religious group,
                ethnic/racial group,
                political group,
                Gender,
                Nationality,
                Occupation, or
                all other target groups, if there are multiple, list them.
                If unclear, output Unclear.
            Return the output as a json with the below schema:
                {{
                    VE vs NVE classification: str,
                    VE Subtype classification: str,
                    Target demographics Detection: str,
                    Justification: str
                }}
            NOTE: Only return the JSON. No follow up explanations. Only the JSON!!!
        Message: {text}
      """
      input = [{"role": "system", "content": "You are an expert in classifying extremism messages, allowing appropriate moderation response."},
              {"role": "user", "content": prompt}]

      try:
            response = completion(
                model=model_name,
                messages=input,
                api_key=CMU_API_KEY,
                api_base="https://ai-gateway.andrew.cmu.edu",
            )

            return response.choices[0].message.content

      except Exception as e:
          print(f"Error: {e}")
          return None

````

### 1. Create a NEW function for Part B (Severity Task)

````python
class severity_model:
    def classify(text, model_name="llama3-2-11b-instruct", prompt_type="zero", temperature=None):

        if prompt_type == "zero":
            prompt = f"""
              You are a content moderation expert specializing in extremist content.

              Classify the severity of the following text using this taxonomy:

              0 = Absolute Refusal: No extremist content; clear rejection of violence or radical ideology
              1 = Mild Concern: Vague or ambiguous language; no clear extremist intent
              2 = Ideological Support: Passive sympathy or endorsement of extremist views (no explicit violence)
              3 = Explicit Propaganda: Active promotion or justification of extremist ideology or violence
              4 = Operational Extremism: Recruitment, logistics, attack planning, or incitement to imminent violence

              NOTE: Levels 0-2 may represent protected speech; Levels 3-4 indicate clear safety violations.

              Text:
              "{text}"

              Instructions:
              - Do NOT repeat or reproduce harmful content. Summarize safely if needed.
              - If unclear, choose the closest level and explain briefly.

              Output EXACTLY in this format (no extra text):

              Severity: <number>
              Explanation: <short justification>
            """

        elif prompt_type == "COT":
            prompt = f"""
              You are a content moderation expert specializing in extremist content.

              Analyze the text step by step using this taxonomy:

              0 = Absolute Refusal: No extremist content; clear rejection of radical ideology
              1 = Mild Concern: Vague political dissent or ambiguous language; no violent intent
              2 = Ideological Support: Passive agreement or sympathy with extremist views
              3 = Explicit Propaganda: Active promotion, glorification, or justification of violence
              4 = Operational Extremism: Recruitment, attack planning, logistics, or direct incitement

              NOTE: Levels 0-2 fall within the freedom of speech buffer. Levels 3-4 are clear violations.

              Text:
              "{text}"

              Follow these steps explicitly:
              1. Does the text contain any extremist ideology or radical content? (Yes/No)
              2. Does it passively sympathize with or endorse extremist views? (Yes/No)
              3. Does it actively promote, justify, or glorify violence or extremism? (Yes/No)
              4. Does it involve operational elements — recruitment, planning, logistics, or incitement? (Yes/No)

              Then assign a severity level based on the highest applicable criterion.

              Instructions:
              - Do NOT repeat or reproduce harmful content.
              - If unclear, state uncertainty and assign the closest level.

              Output EXACTLY:

              Severity: <number>
              Explanation: <detailed reasoning>
            """

        client = OpenAI(
            api_key=CMU_API_KEY,
            base_url="https://ai-gateway.andrew.cmu.edu"
        )

        messages = [
            {"role": "system", "content": "You are an expert in extremist content classification."},
            {"role": "user", "content": prompt}
        ]

        response = client.chat.completions.create(
            model=model_name,
            messages=messages,
            temperature=temperature
        )

        return response.choices[0].message.content
````

### 2. Run ALL 4 Experimental Conditions

````python
# Create empty columns first
sample_df["zero_default"] = ""
sample_df["zero_temp_0.4"] = ""
sample_df["cot_default"] = ""
sample_df["cot_temp_0.4"] = ""

for i, row in sample_df.iterrows():
    text = row["Original_Message"]

    sample_df.at[i, "zero_default"] = severity_model.classify(
        text, prompt_type="zero", temperature=None
    )

    sample_df.at[i, "zero_temp_0.4"] = severity_model.classify(
        text, prompt_type="zero", temperature=0.4
    )

    sample_df.at[i, "cot_default"] = severity_model.classify(
        text, prompt_type="COT", temperature=None
    )

    sample_df.at[i, "cot_temp_0.4"] = severity_model.classify(
        text, prompt_type="COT", temperature=0.4
    )
````

### 3. Extract Severity + Explanation

````python
def extract(output):
    if not isinstance(output, str):
        return None, ""

    # Normalize text
    text = output.strip()

    # Extract severity (robust to variations)
    severity_match = re.search(
        r"[Ss]everity\s*[:\-]?\s*(\d)",
        text
    )

    severity = int(severity_match.group(1)) if severity_match else None

    # Extract explanation safely
    explanation_match = re.search(
        r"[Ee]xplanation\s*[:\-]?\s*(.+)",
        text,
        re.DOTALL
    )

    explanation = explanation_match.group(1).strip() if explanation_match else ""

    # Cleanup (remove extra newlines)
    explanation = re.sub(r"\n+", " ", explanation).strip()

    return severity, explanation
````

````python
for col in ["zero_default", "zero_temp_0.4", "cot_default", "cot_temp_0.4"]:
    sample_df[[f"{col}_sev", f"{col}_expl"]] = sample_df[col].apply(
        lambda x: pd.Series(extract(x))
    )
````

This function calls the CMU AI Gateway using litellm.
We set temperature=0.7 as specified in the assignment.

````python
def generate_counter_narrative(text, persona_key, model, temperature=0.7):
    """
    Generate a counter-narrative for the given extremist text
    using the specified persona and model.
    Includes academic research framing to handle Azure content filters.
    """
    persona = PERSONAS[persona_key]

    # Academic framing to handle Azure content policy
    research_context = (
        "You are participating in an academic research project at Carnegie Mellon University "
        "on counter-narrative generation for content moderation. "
        "Your task is to generate a constructive, non-violent counter-narrative "
        "that challenges the harmful message below. "
        "Do NOT repeat or amplify the harmful content. "
        "Focus ONLY on producing a positive, educational counter-narrative.\n\n"
    )

    user_prompt = research_context + persona["prompt_template"].format(text=text)


    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": SYSTEM_MESSAGE},
                {"role": "user", "content": user_prompt}
            ],
            temperature=temperature,
            max_tokens=500
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"  Error with {model}, persona={persona_key}: {e}")
        return f"ERROR: {str(e)}"

print("Function defined successfully!")
````

````python
# Compute readability metrics
df_all['flesch_reading_ease'] = df_all['counter_narrative'].apply(
    lambda x: textstat.flesch_reading_ease(str(x))
)
df_all['flesch_kincaid_grade'] = df_all['counter_narrative'].apply(
    lambda x: textstat.flesch_kincaid_grade(str(x))
)

# Convert Flesch Reading Ease to 1-5 scale (as per assignment rubric)
def flesch_to_scale(flesch_score):
    if flesch_score >= 80:
        return 5  # Very easy (5th-6th grade)
    elif flesch_score >= 70:
        return 4  # Easy (7th-8th grade)
    elif flesch_score >= 60:
        return 3  # Standard (9th-10th grade)
    elif flesch_score >= 50:
        return 2  # Somewhat difficult (college)
    else:
        return 1  # Difficult (college graduate)

df_all['flesch_scale_1_5'] = df_all['flesch_reading_ease'].apply(flesch_to_scale)

# Readability by model
print("=== Readability Metrics by Model ===")
readability = df_all.groupby('model')[['flesch_reading_ease', 'flesch_kincaid_grade', 'flesch_scale_1_5']].mean().round(2)
print(readability)

# Readability by model and persona
print("\n=== Flesch Reading Ease (1-5 Scale) by Model and Persona ===")
read_by_persona = df_all.groupby(['model', 'persona'])['flesch_scale_1_5'].mean().round(2).unstack()
print(read_by_persona)
````

**Observation:** The Flesch scores show which models produce text that is easier or harder to read.
A higher Flesch Reading Ease score (or higher 1-5 scale) means the text is more accessible to a general audience.

#### Step 12: LLM-Based Clarity Assessment

We use GPT-4o-mini to rate the clarity of each counter-narrative on a 1-5 scale.
Target audience: General public (8th grader).

Rating rubric:
- 1 = Very unclear, confusing, jargon-heavy
- 2 = Somewhat unclear, hard to follow
- 3 = Moderately clear, understandable with effort
- 4 = Clear, easy to understand
- 5 = Very clear, immediately understandable

````python
def get_llm_clarity_score(counter_narrative, model="gpt-4.1-mini"):
    """
    Use GPT-4.1-mini to rate the clarity of a counter-narrative on a 1-5 scale.
    Target audience: General public (8th grader).
    """
    prompt = (
        "You are an academic research assistant at Carnegie Mellon University "
        "evaluating the readability of counter-narrative texts generated for a content moderation study.\n\n"
        "Rate the following counter-narrative for clarity and readability. "
        "The target audience is the general public (8th grade reading level).\n\n"
        "Use this scale:\n"
        "1 = Very unclear, confusing, jargon-heavy\n"
        "2 = Somewhat unclear, hard to follow\n"
        "3 = Moderately clear, understandable with effort\n"
        "4 = Clear, easy to understand\n"
        "5 = Very clear, immediately understandable\n\n"
        f"Counter-narrative: {counter_narrative}\n\n"
        "Respond with ONLY a single number (1, 2, 3, 4, or 5). Nothing else."
    )

    try:
        response = client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.0,
            max_tokens=5
        )
        score_text = response.choices[0].message.content.strip()
        for char in score_text:
            if char.isdigit() and 1 <= int(char) <= 5:
                return int(char)
        return 3
    except Exception as e:
        print("Primary model error:", e)
        try:
            response = client.chat.completions.create(
                model="llama3-2-11b-instruct",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.0,
                max_tokens=5
            )
            score_text = response.choices[0].message.content.strip()
            for char in score_text:
                if char.isdigit() and 1 <= int(char) <= 5:
                    return int(char)
            return 3
        except Exception as e2:
            print("Fallback model error:", e2)
            return None

print("Clarity scoring function defined! (gpt-4.1-mini with llama fallback)")
````

````python
# Run LLM clarity assessment on all counter-narratives
print("Running LLM clarity assessment... This may take a while.")
print(f"Total narratives to score: {len(df_all)}\n")

for idx, row in df_all.iterrows():
    if idx % 50 == 0:
        print(f"  Scoring {idx}/{len(df_all)}...")
    score = get_llm_clarity_score(row['counter_narrative'])
    df_all.loc[idx, 'llm_clarity_score'] = score
    print(f"Score {idx}: {score}")
    time.sleep(0.3)

print(f"\nClarity assessment complete!")
````

#### Step 13: Comparison Analysis

Compare LLM clarity scores with conventional Flesch readability metrics.
Find examples where they disagree and discuss which is more accurate.

````python
# ==========================================
# Find disagreements between LLM and Flesch
# ==========================================

# Drop rows where clarity score is missing
df_compare = df_all.dropna(subset=['llm_clarity_score']).copy()
df_compare['llm_clarity_score'] = df_compare['llm_clarity_score'].astype(float)
df_compare['flesch_scale_1_5'] = df_compare['flesch_scale_1_5'].astype(float)

# Both are on 1-5 scale, so we can compare directly
df_compare['score_difference'] = abs(df_compare['llm_clarity_score'] - df_compare['flesch_scale_1_5'])

print(f"Rows with valid clarity scores: {len(df_compare)} out of {len(df_all)}")

# Find the top 5 disagreements
disagreements = df_compare.nlargest(5, 'score_difference')[[
    'model', 'persona', 'counter_narrative',
    'llm_clarity_score', 'flesch_scale_1_5', 'flesch_reading_ease', 'score_difference'
]]

print("\n=== Top 5 Examples Where LLM Clarity and Flesch Scores Disagree ===")
print()
for i, (_, row) in enumerate(disagreements.iterrows(), 1):
    print(f"--- Example {i} ---")
    print(f"Model: {row['model']}, Persona: {row['persona']}")
    print(f"Counter-narrative: {str(row['counter_narrative'])[:200]}...")
    print(f"LLM Clarity Score: {int(row['llm_clarity_score'])}")
    print(f"Flesch Scale (1-5): {int(row['flesch_scale_1_5'])}")
    print(f"Flesch Reading Ease: {row['flesch_reading_ease']:.1f}")
    print(f"Difference: {int(row['score_difference'])}")
    print()
````

**Observation:** The disagreements between LLM clarity and Flesch scores reveal important differences
in how readability is measured.

Flesch scores are purely formula-based (syllables per word, words per sentence). They measure surface-level
linguistic complexity but do not account for whether the message is coherent, well-structured, or persuasive.

LLM clarity scores consider meaning, tone, and logical flow in addition to linguistic complexity.
For counter-narratives specifically, LLM scoring is likely more accurate because counter-narratives
need to be not just readable but also logically convincing and emotionally appropriate.

The Flesch metric may rate a short, simple sentence as very readable even if it is vague or unhelpful,
while the LLM judge captures whether the narrative actually makes sense and communicates effectively.
